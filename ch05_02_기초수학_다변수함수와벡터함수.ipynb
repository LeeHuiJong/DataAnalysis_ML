### 일변수 스칼라 함수
---

하나의 인풋(=스칼라, 일변수), 하나의 아웃풋(=스칼라, 스칼라함수)

\begin{align}
y = f(x)
\end{align}

#### **예제 01**

\begin{align}
f(x) = x^2+2x+1
\end{align}

* 함수 생성하기

````python
def f1(x): # x는 스칼라
  return x**2 + 2*x +1
f1(-1)
````

* 인풋 정의 (-10~10, 100개)

````python
# 100개의 스칼라 값
xs = np.linspace(-10, 10, 100)
# 100개 스칼라 값에 대한 100개 스칼라 아웃풋
fs = f1(xs)
xs.shape, fs.shape
````

* 2d플롯으로 시각화

````python
fig, ax = plt.subplots(1)
ax.plot(xs, fs )
# 설명
ax.set_title('f(x) acoording to x')
ax.set_xlabel('x') # x축
ax.set_ylabel('f(x)') # y축

x = -1
ax.plot(x, f1(x), marker='o', c='r') # 특정 x에 따른 f(x) 표시
ax.text(x, f1(x), f'({x}, { np.round(f1(x),2) })')
````

* 그래프 설명
  * x가 가질수 있는 값 : 모든 실수
  * f(x)가 가질 수 있는 값: 0 이상의 모든 실수
  * 그래프의 특징 : x가 -1일때 output이 가장 작고 x가 -1에서 멀어질때 output이 커진다.

#### **예제 02**

\begin{align}
f(x) = cosin(x)
\end{align}

* 함수 생성하기

````python
def f2(x):
  return np.cos(x)
````

* 인풋 정의 (-10~10, 100개)

````python
xs = np.linspace(-10,10,100)
fs = f2(xs)
xs.shape, fs.shape
````

* 2d플롯으로 시각화

````python
fig, ax = plt.subplots(1)
ax.plot(xs, fs )
# 설명
ax.set_title('f(x) acoording to x')
ax.set_xlabel('x') # x축
ax.set_ylabel('f(x)') # y축

x = 0
ax.plot(x, f1(x), marker='o', c='r') # 특정 x에 따른 f(x) 표시
ax.text(x, f1(x), f'({x}, { np.round(f1(x),2) })')
````

* 그래프 설명
  * x가 가질수 있는 값 (인풋) : 모든 실수
  * f(x)가 가질 수 있는 값 (아웃풋) : -1 ~ 1 사이의 실수
  * 그래프의 특징 : f(x) 의 값이 -1과 1 밖을 벗어나지 않고 있다. x값의 변화에 따라 반복되는 f(x) 값이 나오고 있다.

### 다변수 스칼라 함수
---

여러개의 인풋(=다변수), 하나의 아웃풋(=스칼라, 스칼라함수)


\begin{align}
y = f(𝐱)\\
\end{align}



#### **예제 03** ★(meshgrid 참조)

\begin{align}
f(x, y) = x+y
\end{align}

* 함수 생성하기

````python
def f3(x, y): # 인풋이 스칼라 두개
  return x+y
f3(2,2)
````

* 인풋 정의 (x, y, 모두 -15~20 값)

````python
xs = np.linspace(-15, 20, 36)
ys = np.linspace(-15, 20, 36)
print(xs.shape, ys.shape)
print(xs)
print(ys)
````

````python
f3(xs, ys)
````

````python
X, Y = np.meshgrid(xs,ys) # xs, ys의 순서쌍 모두 반환
# ★★★ 다변수 일때만 meshgrid 를 해주는건가?
F = f3(X, Y) # X,Y 모든 쌍에 대한 함수 결과 반환
X.shape, Y.shape, F.shape
````

* 3d플롯으로 시각화

````python
# 3차원 플롯
fig = plt.figure(figsize=(9,7))
ax = plt.axes(projection = '3d') # 3차원 플롯
ax.scatter3D(X, Y, f3(X,Y), marker='.', color='k') # x,y,z 축 데이터를 입력 3차원 점 그리기
surf = ax.plot_surface(X, Y, f3(X,Y), cmap='rainbow', alpha=0.7) # 점을 이은 면 그리기 surface를 그리기

# 설명
ax.set_title('f(x,y) acoording to x,y')
ax.set_xlabel('x') # x축
ax.set_ylabel('y') # y축
ax.set_zlabel('f(x,y)') # z축

# 컬러바
fig.colorbar(surf, label='f(x,y)', shrink=0.6)
plt.show()

#x = 0.4
#ax.scatter3D(x, f1(x), marker='o', c='r') # 특정 x에 따른 f(x) 표시
#ax.text(x, f1(x), f'({x}, { np.round(f1(x),2) })')
````

* 그래프 설명
  * x, y가 가질수 있는 값 (인풋) : 모든 실수
  * f(x)가 가질 수 있는 값 (아웃풋) : 모든 실수
  * 그래프의 특징 : 최소값이 없다. x,y의 값이 확장될수록 최소값이 달라지기 때문이다. x,y의 값이 커질수록 f(x,y) 도 같이 커진다.

#### **예제 04**

\begin{align}
f(x, y) = (x-5)^2+(y-5)^2 
\end{align}

* 함수 생성하기

````python
def f4(x,y):
  return (x-5)**2+(y-5)**2
````

* 인풋 정의(x, y 모두 -15~20 사이 값)

````python
xs = np.linspace(-15, 20, 40)
ys = np.linspace(-15, 20, 40)
X, Y = np.meshgrid(xs,ys)
F = f4(X,Y)
X.shape, Y.shape, F.shape
````

* 3d플롯으로 시각화

````python
fig = plt.figure(figsize=(9,7))
ax = plt.axes(projection = '3d') # 3차원 플롯
ax.scatter3D(X, Y, f4(X,Y), marker='.', color='k') # x,y,z 축 데이터를 입력 3차원 점 그리기
surf = ax.plot_surface(X, Y, f4(X,Y), cmap='rainbow', alpha=0.7) # 점을 이은 면 그리기 surface를 그리기

# 설명
ax.set_title('f(x,y) acoording to x,y')
ax.set_xlabel('x') # x축
ax.set_ylabel('y') # y축
ax.set_zlabel('f(x,y)') # z축

# 컬러바
fig.colorbar(surf, label='f(x,y)', shrink=0.6)
plt.show()
````

* 그래프 설명
  * x, y가 가질수 있는 값 (인풋) : 모든 실수
  * f(x,y)가 가질 수 있는 값 (아웃풋) : 0보다 큰 모든 실수 
  * 그래프의 특징 : x와 y 값의 0을 기준으로 벗어 날때 마다 f(x,y)의 값이 점점 커진다.

#### **예제 05**

\begin{align}
f(x_1, x_2) = 50(x_2-x_1^2)^2+(2-x_1)^2
\end{align}

* 함수 생성하기

````python
def f5(x1,x2):
  return 50*(x2 - x1**2)**2 + (2-x1)**2
````

* 인풋 정의 (x: -2 ~ 2 20개, y: -1~3 20개)

````python
x1s = np.linspace(-2, 2, 20)
x2s = np.linspace(-1, 3, 20)
X1, X2 = np.meshgrid(x1s, x2s)
F = f5(X1, X2)
X1.shape, X2.shape, F.shape
````

* 3d플롯으로 시각화

````python
fig = plt.figure(figsize=(9,7))
ax = plt.axes(projection = '3d') # 3차원 플롯
ax.scatter3D(X1, X2, f5(X1, X2), marker='.', color='k') # x,y,z 축 데이터를 입력 3차원 점 그리기
surf = ax.plot_surface(X1, X2, f5(X1, X2), cmap='rainbow', alpha=0.7) # 점을 이은 면 그리기 surface를 그리기

# 설명
ax.set_title('f(x1,x2) acoording to x1,x2')
ax.set_xlabel('x1') # x축
ax.set_ylabel('x2') # y축
ax.set_zlabel('f(x1,x2)') # z축

# 컬러바
fig.colorbar(surf, label='f(x1,x2)', shrink=0.6)
plt.show()
````

* $x_1$, $x_2$의 범위를 둘다 -4부터 4로 변경하여 다시 그려보세요.

````python
x1s = np.linspace(-4, 4, 20)
x2s = np.linspace(-4, 4, 20)
X1, X2 = np.meshgrid(x1s, x2s)
F = f5(X1, X2)
X1.shape, X2.shape, F.shape
````

````python
fig = plt.figure(figsize=(9,7))
ax = plt.axes(projection = '3d') # 3차원 플롯
ax.scatter3D(X1, X2, f5(X1, X2), marker='.', color='k') # x,y,z 축 데이터를 입력 3차원 점 그리기
surf = ax.plot_surface(X1, X2, f5(X1, X2), cmap='rainbow', alpha=0.7) # 점을 이은 면 그리기 surface를 그리기

# 설명
ax.set_title('f(x1,x2) acoording to x1,x2')
ax.set_xlabel('x1') # x축
ax.set_ylabel('x2') # y축
ax.set_zlabel('f(x1,x2)') # z축

# 컬러바
fig.colorbar(surf, label='f(x1,x2)', shrink=0.6)
plt.show()
````

* 그래프 설명
  * x, y가 가질수 있는 값 (인풋) : 모든 실수
  * f(x)가 가질 수 있는 값 (아웃풋) : 0보다 큰 실수
  * 그래프의 특징 : x1이 2, x2 가 4 일때 0의 최소값을 가지고 그 값을 벗어날수록 f(x,y) 값이 점점 커진다. 특히 x1이 x2 보다 f(x1,x2) 값에 영향을 많이 끼친다.

## 일변수 벡터함수
---

하나의 인풋(=스칼라, 일변수), 여러개의 아웃풋(=스칼라, 스칼라함수)




\begin{align}
f(x) = (f_1(x), f_2(x), ..., f_n(x))^T
\end{align}

#### **예제 06**

\begin{align}
f(x) = \begin{pmatrix}
h(x)\\ 
g(x)\\ 
\end{pmatrix}  = \begin{pmatrix}
x^2\\ 
5x\\ 
\end{pmatrix}
\end{align}

* 함수 $f(x)$ 생성하기

````python
def f6(x):
  return np.array([x**2, 5*x])

f6(1)
````

* 인풋 정의 (-10~10 20개)

````python
xs = np.linspace(-10, 10 , 20) # x 값 20개에 따른
hs, gs = f6(xs)
# 인풋    첫번째 벡터(아웃풋)요소 두번째 벡터요소
xs.shape,       hs.shape,            gs.shape
````

* 2d플롯으로 시각화

````python
# 아웃풋(f(x)) 스페이스 그려보기

fig, ax = plt.subplots(1)
ax.plot(gs, hs ) # x축을 g(x), y축을 h(x) 선그래프
# 설명
ax.set_title('f(x) space')
ax.set_xlabel('g(x)') # x축
ax.set_ylabel('h(x)') # y축

scat = ax.scatter(gs, hs, c=xs, cmap='rainbow') # 점그래프
fig.colorbar(scat)
plt.show()
````

* 그래프 설명
  * x가 가질수 있는 값 (인풋) : 모든 실수
  * g(x)가 가질 수 있는 값 (첫번째 아웃풋) : 모든 실수 
  * h(x)가 가질수 있는 값 (두번째 아웃풋): 0보다 큰 실수

#### **예제 07**

\begin{align}
f(t) = \begin{pmatrix}
x(t)\\ 
y(t)\\ 
z(t)\\ \end{pmatrix} = \begin{pmatrix}
sin(6t)\\
\frac{t}{4}\\
\frac{t^2}{2}\\
\end{pmatrix}
\end{align}

* 함수 생성하기

````python
def f7(t):
  return np.array([np.sin(6*t), t/4, t**2/4])

f7(2)
````

* 인풋 정의 (-1~3 200개)

````python
ts = np.linspace(-1, 3, 200)
xs, ys, zs = f7(ts)
xs.shape, ys.shape, zs.shape
````

* 3d플롯으로 시각화

````python
fig = plt.figure(figsize=(9,7))
ax = plt.axes(projection = '3d')
ax.scatter3D(xs, ys, zs) # x축을 x(t), y축을 y(t), z축을 z(t)로 그리는 선그래프
# 설명
ax.set_title('f(t) space')
ax.set_xlabel('x(t)') # x축
ax.set_ylabel('y(t)') # y축
ax.set_zlabel('z(t)') # z축

scat = ax.scatter(xs, ys, zs, c=xs, cmap='rainbow') # 점그래프
fig.colorbar(scat, shrink =0.6)
plt.show()
````

* 그래프 설명
  * t가 가질수 있는 값 (인풋) : 모든 실수
  * x(t)가 가질 수 있는 값 (첫번째 아웃풋) : -1과 1 사이의 실수
  * y(t)가 가질 수 있는 값 (두번째 아웃풋) : 모든 실수
  * z(t)가 가질 수 있는 값 (세번째 아웃풋) : 0보다 큰 모든 실수

## 다변수 벡터함수
---
여러개의 인풋(=스칼라, 일변수), 여러개의 아웃풋(=스칼라, 스칼라함수)

\begin{align}
f(𝐱) = (f_1(𝐱), f_2(𝐱), ..., f_n(𝐱))^T
\end{align}

#### **예제 09**


\begin{align}
f(x,y) = \begin{pmatrix}
x^2\\ 
y^2\\ 
xy\\ \end{pmatrix}
\end{align}

* 함수 생성하기

````python
def f9(x, y):
  return np.array([x**2, y**2, x*y])

f9(1,2)
````

* 인풋 정의 (x,y 모두 -0.5 ~ 1 사이 50개)

````python
xs = np.linspace(-0.5, 1, 50)
ys = np.linspace(-0.5, 1, 50)
# x, y 의 순서쌍
X, Y = np.meshgrid(xs, ys)
A, B, C = f9(X, Y)
X.shape, Y.shape, A.shape, B.shape, C.shape
````

* 3d플롯으로 시각화

````python
fig = plt.figure(figsize=(9,7))
ax = plt.axes(projection = '3d') # 3차원 플롯
ax.scatter3D(A, B, C, marker='.', color='k') # x,y,z 축 데이터를 입력 3차원 점 그리기
surf = ax.plot_surface(A, B, C, alpha =0.5) # 점을 이은 면 그리기 surface를 그리기

# 설명
ax.set_title('f(x,y) space')
ax.set_xlabel('$x^2$') # x축
ax.set_ylabel('$y^2$') # y축
ax.set_zlabel('xy') # z축
````

#### **예제 10**


\begin{align}
f(u,v) = \begin{pmatrix}
u\\ 
v\\ 
1+u^2+\frac{v}{1+v^2}\\ \end{pmatrix}, \ \ \ \ \ 0 \leq u, v \leq1
\end{align}

* 함수 생성하기

````python
def f10(u,v):
  return np.array([u, v, 1+u**2+v/(1+v**2)])

f10(1,1)
````

* 인풋 정의 (u, v 모두 0 ~ 1 사이 50개)

````python
xs = np.linspace(0, 1, 50)
ys = np.linspace(0, 1, 50)

X, Y = np.meshgrid(xs, ys)
A1, B1, C1 = f10(X, Y)
X.shape, Y.shape, A1.shape, B1.shape, C1.shape
````

* 3d플롯으로 시각화

````python
fig = plt.figure(figsize=(9,7))
ax = plt.axes(projection = '3d') # 3차원 플롯
ax.scatter3D(A1, B1, C1, marker='.', color='k') # x,y,z 축 데이터를 입력 3차원 점 그리기
surf = ax.plot_surface(A1, B1, C1, alpha =0.5) # 점을 이은 면 그리기 surface를 그리기

# 설명
ax.set_title('f(u,v) space')
ax.set_xlabel('u') # x축
ax.set_ylabel('v') # y축
ax.set_zlabel('1+u^2+v/(1+v^2)') # z축
````
