### **| 변수 요약**

* 모든 속성의 분포 간단히 그리기

```python
f, axes = plt.subplots(3,3, figsize=(12,12))  # cylinders 는 이산값 weight는 이산형이지만 분포가 연속값을 가지기 때문에 연속형 카테고리에 넣는다.
axes = axes.flatten()                         # model year 이산값 origin 도 범주라서 이산값
for col, ax in zip(df.columns, axes):
  sns.histplot(data = df, x=col, ax=ax)
plt.show()
```

* 자료형 정리하기

```python
nums = ['displacement', 'horsepower','weight','acceleration'] # 연속형 (continuous)
noms = ['cylinders', 'model_year', 'origin'] # 이산형 (nominal)
y = 'mpg' # 예측값
print('연속형 자료', nums)
print('범주/이산자료', noms)
print('예측할 변수', y)
print('추가 전처리가 필요한 속성: name')
```

* 데이터 정리하기
  * 데이터 소개: 1970년 ~ 1980년 초반의 mpg(자동차 연비)를 예측한다. 연비가 높을수록 적은 연료로 많은 거리를 간다.
  * 자료형 소개
    * 연속 자료형
      * mpg : 자동차 연비 (예측값)
      * displacement: 배기량
      * horsepower: 마력
      * acceleration : 가속도
    * 이산 자료형
      * cylinders: 실린더 개수
      * weight: 자동차 무게
      * model_year: 모델의 제조 연도
    * 범주 자료형
      * origin: 제조국가
      * name: 자동차 모델명
  * 모델링을 위한 자료형 구분
    * 연속 -> mpg, displacement, horsepower, acceleration
    * 범주 -> cylinders, weight, model_year
    * 추가 데이터 처리 -> name
    * 예측값 -> mpg

### **| 속성간 관계 시각화**

* `pairplot` 이용해 시각화 하기 

```python
# pairplot은 시간이 오래걸리기 때문에 소량의 샘플을 그리는게 효과적이다.
# df.sample(100) # 데이터 프레임에서 샘플 100개 추출
sns.pairplot(df.sample(100), hue='origin')
```

* `boxplot`으로 범주 속성별 예측값 관계 시각화하기

```python
f, axes = plt.subplots(1,len(noms), figsize=(20,5))
axes = axes.flatten()                         
for col, ax in zip(noms, axes):
  sns.boxplot(data = df, x=col, y='mpg',ax=ax)
plt.show()
# 범주형이기 때문에 boxplot을 사용
# 최신 모델일수록 연비가 좋아진다.
# 일본 모델이 연비가 좋다.
```

* `scatterplot` 이용해 연속 자료형 별 예측값의 관계 시각화하기

```python
# 연속형끼리 비교할때는 scat
# x = 연속자료형 y = mpg
# nums 리스트 변수 그리기
f, axes = plt.subplots(1,len(nums), figsize=(20,5))
axes = axes.flatten()                         
for col, ax in zip(nums, axes):
  sns.scatterplot(data = df, x=col, y='mpg',ax=ax)
plt.show()
```

* `hue` 옵션에 `origin` 추가해서 그리기 

```python
f, axes = plt.subplots(1,len(nums), figsize=(20,5))
axes = axes.flatten()                         
for col, ax in zip(nums, axes):
  sns.scatterplot(data = df, x=col, y='mpg',ax=ax, hue='origin')
plt.show()
```

* `regplot`으로 회귀 시각화

```python
# regression 회귀선을 그려준다
f, axes = plt.subplots(1,len(nums), figsize=(20,5))
axes = axes.flatten()                         
for col, ax in zip(nums, axes):
  sns.regplot(data = df, x=col, y='mpg',ax=ax, marker='.')
plt.show()
```

* `residplot`으로 잔차 시각화

```python
# residplot 잔차
f, axes = plt.subplots(1,len(nums), figsize=(20,5))
axes = axes.flatten()                         
for col, ax in zip(nums, axes):
  sns.residplot(data = df, x=col, y='mpg',ax=ax)
plt.show()
```

* `lmplot`으로 `origin`별 회귀 시각화

```python
for col, ax in zip(nums, axes):
  sns.lmplot(data = df, x=col, y='mpg', hue='origin', col='origin')
  plt.show() # 유럽의 배기량과 mpg는 관계가 있다.
```
