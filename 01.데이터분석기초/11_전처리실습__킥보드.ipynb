## 파일마다 다른 변수의 이름 통일

```python
import pandas as pd
```

```python
data1 = pd.read_csv('2020_rental_dataset_1.csv')  # pd.read_csv 는 괄호 안에 있는 파일명의 파일을 읽기 위한 함수
data2 = pd.read_csv('2020_rental_dataset_2.csv')  # 우선 각각의 데이터를 코랩으로 로드 해줍니다.
data3 = pd.read_csv('2020_rental_dataset_3.csv')
data4 = pd.read_csv('2020_rental_dataset_4.csv')
data5 = pd.read_csv('2020_rental_dataset_5.csv')
```

```python
data1.head()  # y, m/d day time 
```

```python
data2.head()  # y, m,d day t
```

```python
data3.head()  # year, m d day t 
```

```python
data4.head()  # year, m/d day time 
```

```python
data5.head()  # yer, m/d day time 
```

```python
 #각 자료의 연 월 일 요일 시간을 data4 를 기준으로 year, m/d day time 으로 통일 시킨다. count 항목도 counts로 통일시킨다
```

```python
data1=data1.rename({"y":"year"}, axis = 1)
data1.head()
```

```python
data2=data2.rename({"y":"year" , 'm,d':'m/d' , 't':'time'}, axis = 1)
data2.head()
```

```python
data3=data3.rename({"m  d":"m/d" , 't':'time','count':'counts'}, axis = 1)
data3.head()
```

```python
data4=data4.rename({'count':'counts'}, axis = 1)
data4.head()
```

```python
data5=data5.rename({"yer":"year",'count':'counts'}, axis = 1)
data5.head()
```

## 데이터 통합

```python
# 우선 5개의 파일을 통합해서 다루는것이 편리하기 때문에 데이터를 모두 통일 시킨다 
# 여기서 time만 바꾸면 되기 때문에 time 이 안되어있는 data2,data3,data5 를 영문으로 먼저 변형하고 data1,data4의 오타가 있는 monring을 morning으로 고친다.
```

```python
mask = data1['time']=='monring' # 마스크를 사용해 monring을 morning으로 변형한다
```

```python
data1.loc[mask,'time']='morning'
data1.head()
```

```python
mask = data4['time']=='monring'
```

```python
data4.loc[mask,'time']='morning'
data4.head()
```

```python
mask = data2['time']==0    # 시간을 변경할 마스크를 만든다.
mask1= data2['time']==1
mask2= data2['time']==2
```

```python
data2.loc[mask,'time']='morning'
data2.loc[mask1,'time']='noon'
data2.loc[mask2,'time']='evening'
data2.head()
```

```python
mask = data3['time']==0     # data3 에 해당하는 마스크
mask1= data3['time']==1
mask2= data3['time']==2
```

```python
data3.loc[mask,'time']='morning'
data3.loc[mask1,'time']='noon'
data3.loc[mask2,'time']='evening'
data3.head()
```

```python
mask = data5['time']==0     # data5에 해당하는 마스크
mask1= data5['time']==1
mask2= data5['time']==2
```

```python
data5.loc[mask,'time']='morning'
data5.loc[mask1,'time']='noon'
data5.loc[mask2,'time']='evening'
data5.head()
```

```python
#이제 데이터가 모두 동일해졌기 때문에 data1 ~ data5 를 통합한다.
```

```python
data_total= pd.concat([data1, data2, data3, data4, data5], axis=0, ignore_index=True)
data_total
```

## 요일 날씨 값 영문변환

```python
# 시간 값은 이미 변경했기 때문에 동일한 방법으로 요일과 날씨 값을 변경 해준다.
```

```python
mask = data_total['wheather']==0     # data_total 에 날씨에 대한 마스크
mask1= data_total['wheather']==1
mask2= data_total['wheather']==2
```

```python
data_total.loc[mask,'wheather']='serenity'
data_total.loc[mask1,'wheather']='cloudy'
data_total.loc[mask2,'wheather']='rainy'
data_total
```

```python
# 이제 요일 값을 변경해준다.
```

```python
mask = data_total['day']==0     # data_total 에 요일에 대한 마스크
mask1= data_total['day']==1
mask2= data_total['day']==2
mask3= data_total['day']==3
mask4= data_total['day']==4
mask5= data_total['day']==5
mask6= data_total['day']==6
```

```python
data_total.loc[mask,'day']='Sunday'
data_total.loc[mask1,'day']='Monday'
data_total.loc[mask2,'day']='Tuesday'
data_total.loc[mask3,'day']='Wednesday'
data_total.loc[mask4,'day']='Thursday'
data_total.loc[mask5,'day']='Friday'
data_total.loc[mask6,'day']='Saturday'
data_total
```

## 월과 일 분리, 연도 삭제

```python
date_list=data_total['m/d'].str.split(' ')    # 우선 월 일 컬럼을 분리해서 리스트에 담는다.
date_list
```

```python
data_total['month']=date_list.str.get(0)     # 리스트에 있는 요소들을 month 와 date 컬럼을 생성에 각각 담아 준다.
data_total['date']=date_list.str.get(1)
data_total
```

```python
data_total = data_total.drop(['year','m/d'], axis=1)    # 연도는 필요없다고 했으므로 지우고 m/d 컬럼도 나누었으므로 삭제한다.
data_total
```

```python
data_order=['Unnamed: 0','month','date','day','time','wheather','counts']     # 데이터의 순서를 번호, 월, 일, 요일, 시간, 날씨, 대여 횟수로 변경한다
data_total=data_total[data_order]
data_total
```

## 결측값이 있는 행 삭제

```python
data_total.isnull().sum()    # .isnull 함수를 사용하면 couts 항에 9개의 데이터가 결측값을 가지고 있다
```

```python
data_total=data_total.dropna()    # .dropna 함수를 사용해 결측치를 제거한다.
data_total
```

```python
data_total.isnull().sum()  # .dropna 함수 실행후 확인 결과 결측치가 없으므로 제대로 함수가 작동했음을 알 수 있다.
```

## 대여 횟수 실수값을 정수값으로 변환

```python
data_total['counts'] = data_total['counts'].astype('int64')
data_total
```

## 대여수가 많은날 적은날의 특징

```python
mask = data_total['counts']>1000  # 대여횟수가 1000이상인 마스크를 설정하고 값을 뽑아 낸다.
data_total[mask]                  # 이중 가장 많은 날은 1112회 11월 20일 일요일 저녁 맑은 날이다.
                                  # 상관관계를 살펴보면 대여수 1000회 이상의 날씨는 모두 맑음이다. 요일은 모두 일요일이고 주로 저녁과 아침 해가 없는 시간대가 많다.
                                  # 일요일 주말 휴일에 해가 없는 선선한 시간대에 사람들이 모두 많이 외출을 하는것이 보통이기 때문에 이러한 공통적인 특징을 지니는것 같다.
```

```python
mask = data_total['counts']<5 # 대여횟수 5회 미만인 마스크를 만든후에 최소값을 찾는다.
data_total[mask]              # 제일 작은 대여회수는 0회 6월 29일 수요일 오후 비오는 날이다.
                              # 상관관계를 살펴보자면 여름에 비가 오는 날씨로 인해서 덥고 습해지면서 사람들이 외출을 안하게 되고 자연스럽게 대여회수가 적어진 것으로 보인다.
                              # 특히 5회 미만인 날씨와 시간을 보면 다 비오는날의 오후다. 
```
