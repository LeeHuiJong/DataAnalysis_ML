## 05. 범죄의 범주 'Category'는 어떤 값이 있는지 확인하기

## 06. Dates 필드의 데이터타입을 'datetime64' 타입으로 변경하기

```python
train['Dates'] = train['Dates'].astype('datetime64')  # .astype() 함수는 열의 모든 자료들의 데이터를 ()안의 타입으로 변경하는 함수다. .astype({'컬럼명: 타입}) 특정 컬럼을 바꾸려면 이 함수를 쓴다.
```

```python
train.dtypes  # datetime64로 변경후 타입을 확인해 보면 Dates항목의 타입이 바뀐것을 확인할 수 있다.
```

## 07. 연, 월, 일, 요일, 시, 분값을 새로운 컬럼(year, month, day, dayofweek, hour, minute)을 생성해서 할당하기

```python
train['year'] = train['Dates'].dt.year    # year(연도) 컬럼     # .dt 연산자로 Dates에 있는 연,월,일,요일,시,분 값을 각각 뽑아서 새로운 컬럼에 할당시킬수 있다.
train['month'] = train['Dates'].dt.month   # month(월) 컬럼
train['day'] = train['Dates'].dt.day  # day(일) 컬럼
train['dayofweek'] = train['Dates'].dt.dayofweek  # dayofweek(요일) 는 요일을 나타낸다 0,1,2,3,4,5,6 순서대로 월,화,수,목,금,토,일 을 나타낸다.
train['hour'] = train['Dates'].dt.hour # hour(시간) 컬럼
train['minute'] = train['Dates'].dt.minute  # minute(분) 컬럼
```

```python
train  # 항목을 확인해보면 각각의 요소가 잘 할당 되어있는것을 확인할 수 있다.
```

## 08. 범죄가 발생한 경도, 위도 값을 나타내는 X, Y 컬럼의 값을 더한 값을 담은 'X+Y' 컬럼, X, Y의 차이 값을 담은 'X-Y' 컬럼을 생성하기

```python
train['X+Y'] = train['X'] + train['Y']  # 위에서 쓴 새로운 컬럼에 할당하는 함수와 비슷하지만 이번에는 X,Y 값을 더한 값과 뺀 값을 나눠서 항목에 담았다.
train['X-Y'] = train['X'] - train['Y']
```

```python
train # 맨 오른쪽에 X,Y를 더하고 뺀 값의 컬럼이 추가 된 것을 확인할 수 있다.
```

## 09. n_days라는 컬럼을 생성하고, 일별 d-day를 계산하여 입력하기

```python
train['n_days'] = (train['Dates'].dt.date - train['Dates'].dt.date.min()).apply(lambda x: x.days)
test['n_days'] = (test['Dates'].dt.date - test['Dates'].dt.date.min()).apply(lambda x: x.days)
```

## 15. null값이 포함된 데이터 필드가 있는지 확인하기
