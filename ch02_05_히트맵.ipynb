### (1) pivot 테이블 이용한 시계열 데이터 시각화


```
np.random.seed(42)

```

````python
np.random.seed(42)
t = ['오전','오후','저녁']*5
day=[f'2020/01/0{i+1}' for i in range(5) for j in range(3)]
vals = np.random.randint(100,500, len(t))

df=pd.DataFrame([day, t]).T
df.columns=['day','time']
df['count'] = vals
df.head()
````

````python
df=pd.DataFrame([day, t, vals]).T # 같은 데이터 타입을 가진다. 그렇기 때문에 바꾸기 귀찮으니 위에처럼 따로 열을 만들어서 붙여준다. 
df.head()
df.info()
````

````python
df.pivot('day','time','count')
````

````python
        #  행 /   열   / 값
df.pivot('day','time','count').style.background_gradient()
````

### (2) Iris 데이터 변수별 상관관계 분석


```
           꽃받침 길이   꽃받침 너비   꽃의 길이  꽃의 너비
 샘플1      
 샘플2 
 ...
 샘플150
```


```
from sklearn.datasets import load_iris
iris = load_iris()
```


* `.feature_names` : 변수 이름
* `.target_names` : 타겟 이름
* `.data` : 데이터 
* `.target` : 타겟


````python
from sklearn.datasets import load_iris
iris = load_iris()
````

````python
x_names = iris.feature_names
y_names = iris.target_names
print('x names:', x_names)
print('y names:', y_names)
x_data = iris.data
y_data= iris.target
print('x:', x_data.shape)
print('y:', y_data.shape)
````

````python
df=pd.DataFrame(x_data)
df.columns=x_names
df['label']=y_data
df.head()
````

````python
df.corr().style.background_gradient()
````

````python
sns.heatmap(df.corr(), 
            vmin=-1, vmax=1, # 값의 범위. 
            annot=True)
````

````python
plt.figure(figsize=(14,8))
````

### (3) Iris 데이터 종류별 크기 분석



````python
# 라벨이 0/1/2 각각의 평균
df.groupby('label').mean().style.background_gradient()
````

````python
# seaborn 히트맵
sns.heatmap(df.groupby('label').mean(),
            annot=True)
````
