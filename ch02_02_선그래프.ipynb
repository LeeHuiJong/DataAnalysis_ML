````python
a=1

x = np.linspace(-5,5,1000) # -5 에서 5까지 1000개의 데이터로 나눔
def func2(a):  
  y = a*(x**2)
  return y

# 왼쪽
plt.subplot(1,2,1) # (행의개수, 열의개수, 위치)
plt.plot(x,func2(1)) # y= x**2
plt.plot(x,func2(2)) # y= 2x^2

# 오른쪽
plt.subplot(1,2,2)
plt.plot(x,func2(-1)) # y= -x^2
plt.plot(x,func2(-2)) # y= -2x^2
````

### **02. 눈금 공유**
---



````python
t= np.arange(0.01,5,0.01)  # 0.01 부터 5까지 0.01 씩 증가
s1 = np.sin(t*np.pi)
s2 = np.sin(t**np.pi)
s3 = np.exp(-t)

ax1 = plt.subplot(3,1,1)
plt.plot(t,s1)
plt.tick_params('x', labelbottom=False)# 눈금 지우기

plt.subplot(3,1,2, sharex=ax1)  # sharex= x  x를 다른 축으로 정의해 준다.
plt.plot(t,s2)
plt.tick_params('x', labelbottom=False)

plt.subplot(3,1,3, sharex=ax1) # 축공유
plt.plot(t,s3)
````

````python
x = np.linspace(5,-5,1000) # -5 에서 5까지 1000개의 데이터로 나눔
def func2(a):  
  y = np.log(x)
  return y
plt.plot(x,func2(-1))
````

````python
x = np.linspace(0,5,1000) # -5 에서 5까지 1000개의 데이터로 나눔
def func2(a):  
  y = np.exp(x)
  return y
plt.plot(x,func2(1))
````

````python
x = np.linspace(0,5,1000) # -5 에서 5까지 1000개의 데이터로 나눔
def func1(a):  
  y = np.exp(x)
  return y
def func2(a):  
  y = np.log(x)
  return y
plt.plot(x,func1(1))
plt.plot(x,func2(1))
plt.ylim([-100, 100]) 
````

````python
p=np.arange(100)
s4 = np.exp(p)
plt.plot(-s4)
````

````python
t= np.arange(0,10,0.1)  # 0.01 부터 5까지 0.01 씩 증가
s1 = np.exp(t)
s2 = np.exp(t)
s3 = np.exp(-t)
s4 = np.log(t)

ax1 = plt.subplot(4,1,1)
plt.plot(t,s1)
plt.tick_params('x', labelbottom=False)# 눈금 지우기

plt.subplot(4,1,2, sharex=ax1)
plt.plot(t,-s2)
plt.tick_params('x', labelbottom=False)

plt.subplot(4,1,3, sharex=ax1)
plt.plot(t,s3)
plt.tick_params('x', labelbottom=False)

plt.subplot(4,1,4, sharex=ax1)
plt.plot(t,s4)
````

### **02. flights 데이터**

````python
df = sns.load_dataset('flights')
df
````

````python
df.describe()
````

````python
df['passengers'].groupby(df['year']).median()
````

````python
df['passengers'].groupby(df['year']).mean()
````

````python
a=df['passengers'].groupby(df['year']).mean() # 각 연도의 평균과 그룹한 값
plt.plot(a)
````

````python
b=df['passengers'].groupby(df['year']).std() # 각 연도의 표준편차와 그룹한 값
plt.plot(b)
````

````python
c=df['passengers'].groupby(df['year']).min() # 각 연도의 최소값과 그룹한 값
plt.plot(c)
````

````python
d=df['passengers'].groupby(df['year']).max() # 각 연도의 최대값과 그룹한 값
plt.plot(d)
````

````python
e=df['passengers'].groupby(df['year']).median() # 각 연도의 중간값과 그룹한 값
plt.plot(e)
````
